## Classification Accuracy Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison_df["Model"],
    comparison_df["Accuracy"]
)

plt.ylabel("Accuracy")
plt.ylim(0.95, 1.01)

plt.title("Model Accuracy Comparison")

plt.xticks(rotation=15)

plt.tight_layout()

plt.savefig(
    "../results/figures/model_comparison/model_accuracy_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.savefig(
    "../results/figures/model_comparison/model_accuracy_comparison.pdf",
    bbox_inches="tight"
)

plt.show()
plt.close()
```

## Training Time Comparison

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison_df["Model"],
    comparison_df["Training Time (s)"]
)

plt.ylabel("Training Time (seconds)")

plt.title("Model Training Time Comparison")

plt.xticks(rotation=15)

plt.tight_layout()

plt.savefig(
    "../results/figures/model_comparison/model_training_time_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.savefig(
    "../results/figures/model_comparison/model_training_time_comparison.pdf",
    bbox_inches="tight"
)

plt.show()
plt.close()
```

# Comparative Analysis

All four models demonstrated excellent predictive performance on the selected transcriptomic dataset, achieving perfect classification accuracy on the held-out test set.

The classical machine learning algorithms (Logistic Regression, Random Forest, and Support Vector Machine) produced identical classification performance, indicating that the selected transcriptomic biomarkers possess strong discriminative capability regardless of the classification algorithm employed.

The Artificial Neural Network also achieved perfect predictive performance. Although deep learning models are capable of capturing highly complex nonlinear relationships, the absence of any measurable improvement over the classical models suggests that the feature engineering pipeline had already transformed the data into a highly separable feature space.

Among the evaluated models, Support Vector Machine required the shortest training time while maintaining perfect predictive performance. Logistic Regression provided excellent interpretability with similarly high accuracy, making it an attractive baseline classifier for biomedical applications where model transparency is important.

Random Forest offers additional robustness to noisy biological data and provides feature importance estimation, which may be beneficial for future biomarker discovery studies.

The Artificial Neural Network required substantially longer training time than the classical approaches while providing no improvement in predictive accuracy on the current dataset. This observation highlights that increased model complexity does not necessarily translate into superior performance when informative feature engineering has already been performed.

# Conclusion

This comparative evaluation demonstrates that careful transcriptomic feature engineering can substantially simplify the downstream classification problem.

While all evaluated algorithms achieved perfect predictive performance, simpler classical machine learning models provided comparable accuracy with lower computational cost and greater interpretability.

For this transcriptomic cancer classification task, Support Vector Machine emerged as the most computationally efficient model, whereas Logistic Regression offered the highest degree of interpretability. The Artificial Neural Network confirmed the robustness of the selected feature set but did not provide additional predictive benefit relative to the classical approaches.

These findings indicate that appropriate biological feature selection is likely to contribute more to predictive performance than increasing model complexity for this dataset.
