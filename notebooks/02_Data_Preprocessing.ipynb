# Data Preprocessing

## Objective

This notebook performs preprocessing of the raw TCGA gene expression dataset before exploratory analysis and downstream machine learning.

The preprocessing workflow includes:

- Loading raw gene expression and sample labels
- Checking dataset structure
- Validating missing values
- Ensuring sample-label alignment
- Separating expression features and classification labels
- Saving cleaned datasets for downstream analysis

The processed datasets generated here are used in subsequent EDA and feature engineering workflows.

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
expression_df = pd.read_csv(
    "../data/raw/data.csv",
    index_col=0
)

expression_df.head()
```

```python
labels_df = pd.read_csv(
    "../data/raw/labels.csv"
)

labels_df.head()
```

```python
print("Expression matrix shape:", expression_df.shape)
print("Labels shape:", labels_df.shape)
```

```python
print(
    "Missing values in expression matrix:",
    expression_df.isnull().sum().sum()
)

print(
    "Missing values in labels:",
    labels_df.isnull().sum().sum()
)
```

```python
print(expression_df.index[:5])
print(labels_df["Unnamed: 0"].head())
```

```python
sample_alignment = (
    expression_df.index == labels_df["Unnamed: 0"]
)

print("All samples aligned:", sample_alignment.all())
```

```python
X = expression_df.copy()

y = labels_df["Class"]
```

```python
print("Feature matrix:", X.shape)
print("Labels:", y.shape)
```

```python
processed_path = Path("../data/processed")
processed_path.mkdir(exist_ok=True)
```

```python
X.to_csv(
    processed_path / "clean_expression_matrix.csv"
)

labels_df.to_csv(
    processed_path / "clean_labels.csv",
    index=False
)

print("Processed datasets saved.")
```
